```python
import matplotlib.pyplot as plt
import numpy as np


models = ['Llama3.1_8b_AD', 'Mistral_7b_vO.3_AD', 'Qwen2.5_7b_AD']


train_loss = [1.0641371938275945, 0.8804376355527965, 1.0696223059231664]
train_runtime = [19595.1973, 22655.8497, 18998.8032]
train_samples_per_second = [10.791, 9.333, 11.13]
train_steps_per_second = [0.169, 0.146, 0.174]


plt.rcParams['font.family'] = 'Arial'


fig, axs = plt.subplots(1, 4, figsize=(20, 5),dpi=600)


axs[0].plot(models, train_loss, marker='o', color='tab:blue', label='Train Loss')
axs[0].set_title('Train Loss')
axs[0].set_xlabel('Models')
axs[0].set_ylabel('Train Loss')
axs[0].grid(True)


axs[1].plot(models, train_runtime, marker='o', color='tab:orange', label='Train Runtime')
axs[1].set_title('Train Runtime')
axs[1].set_xlabel('Models')
axs[1].set_ylabel('Train Runtime')
axs[1].grid(True)


axs[2].plot(models, train_samples_per_second, marker='o', color='tab:green', label='Train Samples per Second')
axs[2].set_title('Train Samples per Second')
axs[2].set_xlabel('Models')
axs[2].set_ylabel('Train Samples per Second')
axs[2].grid(True)


axs[3].plot(models, train_steps_per_second, marker='o', color='tab:red', label='Train Steps per Second')
axs[3].set_title('Train Steps per Second')
axs[3].set_xlabel('Models')
axs[3].set_ylabel('Train Steps per Second')
axs[3].grid(True)

plt.tight_layout()
plt.savefig("XX.svg",dpi=600)

plt.show()
```
